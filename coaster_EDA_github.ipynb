### Understanding Missing Values

```python
df.isna().sum().sort_values(ascending=False) # number of missing values per column
```

```python
msno.matrix(df, labels=True)
```

```python
msno.bar(df[df.isna().sum().sort_values(ascending=False).index]) # msno.bar(df) doesnt order by greatest
```

```python
sns.heatmap(df.isnull(), cbar=False) # not all columns
```

```python
df.isnull()
```

```python
df[df.isnull().any(axis=1)] # rows with at least one missing value
```

```python
df[df.isnull().sum(axis=1) >= df.shape[1]*0.4] # rows with at least 40% missing values
```

### EDA with Visualizations 

#### Bar Charts (catergorical and numerical)

```python
df['Location'].unique()
```

```python
# to see outputs in columns with less than 10 unique outputs
for column in df.columns:
    if len(df[column].unique()) < 10:
        print(f"Unique values in '{column}': {df[column].unique()}")
```

```python
# to see unique values/outputs in all OBJECT type columns
# sometimes doesn't work as intended since some object columns have so many unique values
# like doesn't work here

for column in df.columns:
    if df[column].dtype == 'object':
        unique_values = df[column].unique()
        print(f"Unique values in '{column}': {unique_values}")
```

```python
df["year_introduced"].value_counts().reset_index()#.sort_index()
```

Pandas plotting (less customizable, have to be more specific, more common for EDA)

```python
df["year_introduced"].value_counts() \
    .head(10) \
    .plot(kind='bar', 
          title = 'Top 10 Years for Coaster Introductions',
          xlabel='Year Introduced',
          ylabel='Count')
```

```python
#same as above but with ax object to modify after plotting
ax = df["year_introduced"].value_counts() \
    .head(10) \
    .plot(kind='bar', title = 'Top 10 Years for Coaster Introductions')
    
ax.set_xlabel('Year Introduced')
ax.set_ylabel('Count')
```

Matplotlib plotting

```python
introduced_yr_counts = df["year_introduced"].value_counts().head(10)
introduced_yr_counts.reset_index()
```

```python
df["year_introduced"].value_counts().head(10).index #.astype(str)
```

```python
df["year_introduced"].value_counts().head(10).values
```

```python
plt.bar(introduced_yr_counts.index.astype(str), introduced_yr_counts.values)
plt.title('Top 10 Years Coasters Introduced')
plt.xlabel('Year Introduced')
plt.ylabel('Count')
plt.show()
```
